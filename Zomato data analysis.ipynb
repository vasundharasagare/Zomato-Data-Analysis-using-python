# Zomato data analysis project

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Zomato data .csv")
```

```python
df.head()
```

# Convert the datatype of column rate

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0];
    return float(value)

df['rate']= df['rate'].apply(handleRate)
print(df.head())
```

```python
df.info()
```

# Type of resturant

```python
df.head()
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel('type of resturant')
```

```python
a=df.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':a})
plt.plot(result,c="blue", marker="o")
plt.xlabel("Type of resturant", c="red", size=20)
plt.ylabel("Votes", c="red", size = 20)
```

```python
plt.hist(df['rate'],bins =5)
plt.title('ratings distribution')
plt.show()
```

# Average order spending by couple

```python
cd = df["approx_cost(for two people)"]
sns.countplot(x=cd)
```

# which mode receives maximum rating

```python
plt.figure(figsize = (6,6))
sns.boxplot(x= 'online_order', y ='rate',data = df)
```

```python
piovt_table = df.pivot_table(index = 'listed_in(type)', columns = 'online_order', aggfunc = 'size', fill_value = 0)
sns.heatmap(piovt_table, annot = True, cmap ='YlGnBu', fmt='d')
plt.title('Heatmap')
plt.xlabel('Online Order')
plt.ylabel('Listed In(Type)')
plt.show()
```
